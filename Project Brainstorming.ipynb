Acquiring and storing a sizeable database of podcasts is prohibitively difficult. 

```python
import numpy as np

download_speed = 1*(60^2) #mb/hr
podcast_hour_size = 50/1000 #gb/hour recording
desired_podcast_number = 100000 #number of podcasts for dataset
eps_per_podcast = 10 #number of hours
sample_time_per_ep = 6/60 #in hours

#How much data will this be?
total_dataset_size = desired_podcast_number*(sample_time_per_ep*podcast_hour_size*eps_per_podcast) #num podcasts * (time/ep)*(ep)*(size/hour)

print(total_dataset_size)

#How long until this data is downloaded
download_time = total_dataset_size*download_speed*(1/1000)#size * speed * mb/gb

print(download_time)
```
